```python
SyrList = ('ܐܒܓܕܗܘܙܚܛܝܟܠܡܢܣܥܦܨܩܪܫܬ܀')
syriac_pattern = re.compile(r'([ܐܒܓܕܗܘܙܚܛܝܟܠܡܢܣܥܦܨܩܪܫܬ]+\W+)+(?=([a-zA-Z0-9])|.$)', re.UNICODE|re.MULTILINE)
for match in re.finditer(syriac_pattern, myfile):
    print(match.group())
```

```python
myfile = open('Wright Florence, 925-926.txt', encoding = "utf8") 
for line in myfile.readline():
    try:
        match = re.search(syriac_pattern, line).group()
    except:
        transformed = open("transformed_Wright.txt", "a")
        transformed.write(line)
        continue
    else: 
        syriac = open('Wright_Florence_syriac 925-926.txt', encoding = "utf8") 
        syriac_repl = syriac.readline()
        print('this is the original text: '+ match)
        print('this is the replacement: ' + syriac_repl)
        re.sub(match, syriac_repl, line)
        transformed.write(line)
        print('this is the new text: ' + transformed.readline())
        
    
            
   
```
