```python
Trans= doc.paragraphs[0].text
type(Trans)
```

```python
SyrList = ('ܐܒܓܕܗܘܙܚܛܝܟܠܡܢܣܥܦܨܩܪܫܬ̣̱̤̇̈ܲ̄̄̇̈()؟܀.:')
len(SyrList)
```

```python
EngList = ('abgdhwz;jyklmnsipxqrvtP:IZOL!K$%()?*.|')
len(EngList)
```

```python
table = Trans.maketrans(EngList, SyrList)
type(table)
```

```python
Conv = Trans.translate(table)
print(Conv)
len(Conv)
```

```python
NewDoc = docx.Document()
NewDoc.add_paragraph(Conv)
run=NewDoc.paragraphs[0].add_run()
print(run.font.rtl)
run.font.rtl=True
print(run.font.rtl)
font=NewDoc.paragraphs[0].style.font
print(font.name)
font.name='Times New Roman (Headings CS)'#this is not working, try applying it from the beginning of the document
font1=NewDoc.paragraphs[0].style.font
print(font1.name)
font1.name='Estrangelo Edessa'
```
